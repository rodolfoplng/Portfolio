# Simple Linear Regression using salary dataframe

#### from YBI Foundation with edits

```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot( x = df['Experience Years'], y = df['Salary'])
sns.regplot(x=df['Experience Years'], y=df['Salary'])

plt.xlabel('Experience Years')
plt.ylabel('Salary')

plt.show()
```

```python
# define target (y) and features (X)
y = df['Salary']
X = df[['Experience Years']]
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# fit model
lr.fit(X_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
# predict model
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
# model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
sns.scatterplot(x = y_test, y = y_pred, alpha=.5)

plt.title('Salary Predictions vs Truth using Linear Regression')
plt.xlabel('Ground Truth')
plt.ylabel('Predictions')

plt.show()
```
